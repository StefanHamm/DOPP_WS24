```python
import re
```

```python
def drop_columns_with_less_than_3_digits(in_df):
    columns_to_drop = []
    for col in in_df.columns:
        if len(re.findall(r'\d', col)) < 3:
            columns_to_drop.append(col)

    out_df = in_df.drop(columns=columns_to_drop)
    return out_df
```

```python
df_score_without_averages = drop_columns_with_less_than_3_digits(df_score)
```

```python
df_score_without_averages.head()
```

```python
corr_matrix = df_score_without_averages.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix,)
plt.title("Correlation Matrix")
plt.show()
```
